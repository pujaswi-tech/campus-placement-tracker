# Filtering Data with many conditions(AND)

```python
import pandas as pd
```

```python
dataset= pd.read_csv("Campus_Placement_Tracker_500.csv")
```

```python
dataset.head(10)
```

```python
mask1 = dataset.Gender == "Female"
mask1.head()
```

```python
CTC = dataset["CTC (LPA)"]
```

```python
mask2 = dataset["CTC (LPA)"] > 10.00
mask2.head()
```

```python
(mask1 & mask2).head()
```

```python
female_dream = dataset.loc[mask1 & mask2,["Branch","Company","CTC (LPA)","Job Role"]]
female_dream.head()
```

```python
female_dream.info()
```

```python
female_dream.describe()
```

```python
dataset.describe()
```

# Filtering Data by many conditions(OR)

```python
(mask1|mask2).head()
```

```python
dataset.loc[mask1|mask2]
```

```python
female_or_dream = dataset.loc[mask1|mask2,["Branch","Company","CTC (LPA)","Job Role"]]
```

```python
female_or_dream.head()
```

```python
female_or_dream.info()
```

```python
female_or_dream.describe()
```
