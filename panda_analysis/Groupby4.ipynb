```python
import pandas as pd
```

```python
dataset= pd.read_csv("Campus_Placement_Tracker_500.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
Jobs_per_branch = dataset.groupby("Branch")["CTC (LPA)"].count()
Jobs_per_branch
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
Jobs_per_branch.plot(kind = "bar",figsize =(14,8),fontsize = 14)
plt.xlabel("Branch",fontsize=13)
plt.ylabel("No of Jobs",fontsize=13)
plt.title("Placement statistics branchwise",fontsize = 16)
plt.show()
```
