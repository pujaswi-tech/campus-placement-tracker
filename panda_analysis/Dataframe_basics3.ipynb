# Using apply(),map(),applymap()

```python
import pandas as pd
```

```python
dataset= pd.read_csv("Campus_Placement_Tracker_500.csv")
```

```python
dataset["Gender Short"] = dataset["Gender"].map({"Male": "M", "Female": "F"})
```

```python
dataset
```

```python
def classify_ctc(ctc):
    if ctc > 20:
        return "Super Dream"
    elif ctc > 10:
        return "Dream"
    else:
        return "Non-Dream"

dataset["CTC Category"] = dataset["CTC (LPA)"].apply(classify_ctc)
```

```python
dataset.head()
```

```python
def full_name(row):
    return f"{row['First Name']} {row['Last Name']}"

dataset["Full Name"] = dataset.apply(full_name, axis=1)
```

```python
dataset.head()
```

```python
dataset["First Name"] = dataset["First Name"].map(str.upper)
dataset["Last Name"] = dataset["Last Name"].map(str.upper)
```
