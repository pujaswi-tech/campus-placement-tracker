# Creating Columns based on Other Columns

```python
import pandas as pd
```

```python
dataset= pd.read_csv("Campus_Placement_Tracker_500.csv")
```

```python
dataset.head(10)
```

```python
dataset["Dream"] = dataset["CTC (LPA)"] > 10.00
```

```python
dataset["Dream Category"] = dataset["CTC (LPA)"].apply(lambda x: "Dream" if x > 10.00 else "Non-Dream")
```
